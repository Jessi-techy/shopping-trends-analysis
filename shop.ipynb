## 1 What is the overall distribution of customer ages in the dataset?

```python
#summary statistics of the Age column
age_distribution=data['Age'].describe()
age_distribution
```

```python
data['Gender'].unique()
```

```python
# Defining age groups
def age_group(age):
    if age <= 20:
        return 'Teen'
    elif 21 <= age <= 35:
        return 'Young Adult'
    elif 36 <= age <= 55:
        return 'Middle-Aged Adult'
    else:
        return 'Older Adult'

# Adding an age group column to the dataset
data['Age Group'] = data['Age'].apply(age_group)

# Counting age group distribution across categories
age_group_distribution = data.groupby('Category')['Age Group'].value_counts(normalize=True).unstack()

# Summing up to get a percentage distribution for each age group across all categories
age_group_totals = age_group_distribution.sum(axis=0)

# Plotting a pie chart
plt.figure(figsize=(7, 7))
plt.pie(age_group_totals, labels=age_group_totals.index, autopct='%1.1f%%', startangle=140, colors=plt.cm.Paired.colors)
plt.title("Age Group Distribution Across Categories")
plt.show()
```

## 10 How does the frequency of purchases vary across different age groups?

```python
data['Age_category'] = pd.cut(data['Age'], bins= [0,15, 18 , 30 , 50 , 70] , labels= ['child' , 'teen' , 'Young Adults' ,'Middle-Aged Adults'
                                                                                             , 'old'] )
data[['Age','Age_category']]
```

```python
# Group by Age Group and Frequency of Purchases
age_purchase_frequency = data.groupby(['Age_category', 'Frequency of Purchases'],observed=False).size().unstack(fill_value=0)

# Plotting a stacked bar chart
age_purchase_frequency.plot(kind='bar', stacked=True, figsize=(10, 6), colormap='viridis')

# Add labels and title
plt.title('Frequency of Purchases Across Age Groups')
plt.xlabel('Age Group')
plt.ylabel('Number of Purchases')
plt.legend(title='Frequency of Purchases')
plt.xticks(rotation=0)
plt.show()
```
